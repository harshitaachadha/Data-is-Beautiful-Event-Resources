# Let's create our Word Cloud!

```python
# Creating the netflix brand palette in seaborn

sns.palplot(['#221f1f', '#b20710', '#e50914','#f5f5f1'])
```

```python
#Importing specific libraries

from wordcloud import WordCloud
from wordcloud import STOPWORDS
import random
from PIL import Image
import matplotlib

#You might have to install wordcloud using pip
```

```python
# Custom colour map based on Netflix palette

cmap = matplotlib.colors.LinearSegmentedColormap.from_list("", ['#221f1f', '#b20710'])
text = str(list(df['title'])).replace(',', '').replace('[', '').replace("'", '').replace(']', '').replace('.', '')
```

### Basic Word Cloud

```python
wordcloud = WordCloud(background_color = 'white',colormap=cmap, max_words = 500,
                      stopwords=STOPWORDS, random_state=5).generate(text)

plt.figure( figsize=(5,5))
plt.imshow(wordcloud, interpolation = 'bilinear')
plt.axis('off')
plt.tight_layout(pad=0)
plt.show()
```

## Image Word Cloud

```python
mask = np.array(Image.open('Netflix-Logo.png'))

wordcloud = WordCloud(background_color = 'white', width = 500,  height = 200,
                      colormap=cmap, max_words = 150, 
                      stopwords=STOPWORDS, random_state=5,
                      mask = mask).generate(text)

plt.figure( figsize=(5,5))
plt.imshow(wordcloud, interpolation = 'bilinear')
plt.axis('off')
plt.tight_layout(pad=0)
plt.show()
```
